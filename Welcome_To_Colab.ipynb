```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
working_dir_path = '/content'
df=pd.read_csv(working_dir_path + '/Global YouTube Statistics.csv', encoding='latin1')
```

```
category_subscriber_mean=df.groupby('category')['subscribers'].mean()
category_subscriber_mean
```

```
category_subscriber_mean.sort_values(ascending=False, inplace=True)
```

```
category_subscriber_mean.head().plot(kind='bar')
plt.title("category_subscriber_mean")
plt.xlabel("category")
plt.ylabel("mean")
plt.show()
```

```
channel_type_counts=df["channel_type"].value_counts()
channel_type_counts
```

```
channel_type_counts.head().plot(kind='pie', autopct="%1.1f%%")
plt.title("channel type distribution")
plt.show()
```

```
plt.scatter(df["subscribers"], df["video views"])
plt.title('subscribers vs video views')
plt.xlabel('subscribers')
plt.ylabel('Video views')
plt.show()
```

```
country_subscribers_total=df.groupby('Country')["subscribers"].sum().sort_values(ascending=False)
country_subscribers_total
```

```
country_subscribers_total.head().plot(kind='bar')
plt.title('subs by country')
plt.xlabel('Country')
plt.ylabel('total subs')
plt.xticks(rotation=90)
plt.show()
```

```
sns.boxplot(x='channel_type', y='highest_monthly_earnings', data=df)
plt.title('distribution')
plt.xticks(rotation=90)
plt.show()
```

```
#unstack moves the inner level index level (Country ) into columns means row indexes of country becomes columns
df.groupby(['category','Country'])['subscribers'].sum().head().unstack().plot(kind='bar', stacked=True)
plt.title('subs break down by country and category')
plt.xlabel('category')
plt.ylabel('Subs')
plt.legend(title='Country', bbox_to_anchor=(1.05,1),loc='upper left')
plt.xticks(rotation=90)
plt.show()
```

```
plt.scatter(df['Gross tertiary education enrollment (%)'], df['subscribers'], alpha=0.5)
plt.title('scatter plot os subs vs tertiary')
plt.xlabel('tertiary')
plt.ylabel('subs')
plt.show()
```

```
df.groupby('channel_type')['Unemployment rate'].mean().sort_values(ascending=False).plot(kind='bar')
plt.title('unemployment rate by channel type')
plt.xlabel('channel_type')
plt.ylabel('unemp rate')
plt.xticks(rotation=90)
plt.show()
```

```
df.groupby(['created_year','channel_type'])['subscribers'].mean().unstack().tail().plot(kind='bar',width=1)
plt.title('subs by created year and channel type')
plt.xlabel('created year')
plt.ylabel('subs')
plt.xticks(rotation=45)
plt.legend(title='Channel_type', bbox_to_anchor=(1.05,1), loc='upper left')
plt.show()
```

```
df.columns
```

```
df.groupby(['created_year','Country'])['highest_yearly_earnings'].mean().unstack().head().plot(kind='line', marker='o')
plt.title('trends')
plt.xlabel('created_year')
plt.ylabel('yearly earning')
plt.legend(title='Country', bbox_to_anchor=(1.05,1), loc='upper left')
plt.xticks(rotation=45)
plt.show()
```
